```python
drinks.head()
```

## string method in pandas

```python
orders = pd.read_table('https://bit.ly/chiporders')
```

```python
orders.head()
```

```python
## change a column in upper case
orders.item_name.str.upper().head()
```

```python
orders.item_name.str.contains('chicken')
```

```python
orders[orders.item_name.str.contains('Chicken')]
```

```python
## remove the [] character from strings
```

```python
orders.choice_description.str.replace('[', '').head()
```

```python
## replace more than one character
```

```python
orders.choice_description.str.replace('[', '').str.replace(']','').head()
```

## how to change datatype of a data series

```python
drinks.head()
```

```python
drinks.dtypes
```

```python
drinks['beer_servings'] = drinks.beer_servings.astype(float).head()
```

```python
drinks.dtypes
```

```python
## to do math on a string column
```

```python
orders.head()
```

```python
## here item price has $- remove it for calculations
```

```python
orders.dtypes
```

```python
orders.item_price.str.replace('$','').head()
```

```python
##But its still string
orders.dtypes
```

```python
## caste it to a float
orders['item_price' ] = orders.item_price.str.replace('$','').astype(float).mean()
```

```python
orders.dtypes
```

```python
orders.describe()
```

## Machine Learning

```python
import pandas as pd
```

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
feature_cols = ['Pclass','Parch']
```

```python
X = train.loc[:, feature_cols]
```

```python
X.shape
```

```python
y = train.loc[:, 'Survived']
```

```python
y.shape
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
logreg = LogisticRegression()
```

```python
logreg.fit(X,y)
```

```python
test = pd.read_csv('http://bit.ly/kaggletest')
```

```python
test.head()
```

```python
test.shape
```

```python
X_new = test.loc[:, ['Pclass', 'Parch']]
```

```python
X_new.shape
```

```python
new_pred_class = logreg.predict(X_new)
```

```python
new_pred_class
```

```python
pd.DataFrame({'PassengerId':test.PassengerId,  'Survived': new_pred_class}).set_index('PassengerId').head()
```

## dummy variables

```python
train = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
train.head()
```

```python
## make a new column for Sex column
```

```python
train['Sex_Mail'] = train.Sex.map({'female':0, 'male':1})
```

```python
train.head()
```

```python
pd.get_dummies(train.Sex).head()
```

```python
## create dummies for Embarked
```

```python
train.Embarked.value_counts()
```

```python
Embarked_dummies = pd.get_dummies(train.Embarked, prefix='Embarked').loc[:, ['Embarked_S', 'Embarked_C']]
```

```python
train = pd.concat([train , Embarked_dummies], axis = 1)
```

```python
train.head()
```

```python
## get dummies directly from DF
```

```python
dumm =pd.get_dummies(train, columns=['Sex', 'Embarked'], drop_first=True)
```
